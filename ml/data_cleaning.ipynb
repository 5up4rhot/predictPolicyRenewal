```python
import pandas as pd
import numpy as np
```

```python
desc = pd.read_csv('description.csv', sep=';', encoding='1251', index_col=0)
desc
```

```python
data = pd.read_csv('data.txt', sep=';')
data.head()
```

```python
# variable for pre-modeling
df = data.copy()
df.info()
```

```python
for col_name in df.columns:
    if df[col_name].dtypes == 'object':
        n_unique = len(df[col_name].unique())
        print(f"Feature '{col_name}' has {n_unique} unique categories")
```

```python
# devide CLIENT_REGISTRATION_REGION in 6 groups: Санкт-Петербург, Москва, Московская, Ленинградская, Other
print(df['CLIENT_REGISTRATION_REGION'].value_counts().sort_values(ascending=False).head(10))
```

```python
df['CLIENT_REGISTRATION_REGION'] = [x if x in ['Санкт-Петербург', 'Москва', 'Московская', 'Ленинградская'] else 'Other' for x in df['CLIENT_REGISTRATION_REGION']]
df['CLIENT_REGISTRATION_REGION'].value_counts()
```

```python
# devide POLICY_SALES_CHANNEL in groups with more than 500 policies and Other
channel_count = df['POLICY_SALES_CHANNEL'].value_counts()
print(channel_count.sort_values(ascending=False).head(15))
```

```python
df['POLICY_SALES_CHANNEL'] = [x if channel_count[x] > 500 else 'Other' for x in df['POLICY_SALES_CHANNEL']]
df['POLICY_SALES_CHANNEL'].value_counts()
```

```python
# devide VEHICLE_MAKE in groups with more than 100 cars and Other
make_count = df['VEHICLE_MAKE'].value_counts()
print(make_count.sort_values().head())
```

```python
df['VEHICLE_MAKE'] = [x if make_count[x] > 100 else 'Other' for x in df['VEHICLE_MAKE']]
```

```python
# devide VEHICLE_MODEL in groups with more than 100 cars and Other
model_count = df['VEHICLE_MODEL'].value_counts()
print(model_count.sort_values())
```

```python
df['VEHICLE_MODEL'] = [x if model_count[x] > 30 else 'Other' for x in df['VEHICLE_MODEL']]
```

```python
df['VEHICLE_MODEL'].value_counts(ascending=False)
```

```python
group_count = df['POLICY_SALES_CHANNEL_GROUP'].value_counts()
print(group_count)
```

```python

df['POLICY_SALES_CHANNEL_GROUP'] = [x if group_count[x] > 3000 else 'Other' for x in df['POLICY_SALES_CHANNEL_GROUP']]
df['POLICY_SALES_CHANNEL_GROUP'].value_counts()
```

```python
# looking for cat features
df.info()
```

```python
for_encoding = ['POLICY_SALES_CHANNEL', 'POLICY_SALES_CHANNEL_GROUP', 'POLICY_BRANCH', 'VEHICLE_MAKE', 'VEHICLE_MODEL', 'INSURER_GENDER', 'POLICY_CLM_N',
'POLICY_CLM_GLT_N', 'POLICY_PRV_CLM_N', 'POLICY_PRV_CLM_GLT_N', 'CLIENT_REGISTRATION_REGION']

# outliers in: POLICY_MIN_DRIVING_EXPERIENCE (2015 years), VEHICLE_ENGINE_POWER (2000hp), VEHICLE_SUM_INSURED (0rub)
df[df.columns[~df.columns.isin(for_encoding)]].describe()
```

```python
# at some observations driving experience representing first driving year
df['POLICY_MIN_DRIVING_EXPERIENCE'].sort_values(ascending=False).head(5)
```

```python
df['POLICY_MIN_DRIVING_EXPERIENCE'] = [2021 - x if x > 1900 else x for x in df['POLICY_MIN_DRIVING_EXPERIENCE']]
df['POLICY_MIN_DRIVING_EXPERIENCE'].describe()
```

```python
# engine hp can't be 0 or more than ~700
df = df.drop(df[(df['VEHICLE_ENGINE_POWER']>700) | (df['VEHICLE_ENGINE_POWER']<=0)].index)
df['VEHICLE_ENGINE_POWER'].sort_values(ascending=False)
```

```python
# insurance sum can't be 0
df = df.drop(df[df['VEHICLE_SUM_INSURED']<=0].index)
df['VEHICLE_SUM_INSURED'].describe()
```

```python
# drop bad POLICY_CLM_N
df = df.drop(index=df[df['POLICY_CLM_N'] == 'n/d'].index)
```

```python
# drop bad POLICY_YEARS_RENEWED_N
df = df.drop(index=df[df['POLICY_YEARS_RENEWED_N'] == 'N'].index)
```

```python
# cast POLICY_YEARS_RENEWED_N to integer 
df['POLICY_YEARS_RENEWED_N'] = pd.to_numeric(df['POLICY_YEARS_RENEWED_N'], downcast='integer')
df['POLICY_YEARS_RENEWED_N'].describe()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
bad = ['POLICY_ID', 'POLICY_END_MONTH', 'POLICY_INTERMEDIARY']

df = df.drop(bad, axis=1)
df.to_csv('clean_data.csv')
df.info()
```

```python
encoded_df = pd.concat([pd.get_dummies(df[for_encoding]), df.drop(for_encoding, axis=1)], axis=1)
encoded_df.shape
```

```python
encoded_df.to_csv('encoded_data.csv')
```
